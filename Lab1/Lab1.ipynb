```python
import pandas

f = pandas.read_csv("plant-classification.csv", header = None, names = ["habitat", "colour", "type of leaf", "leaf width", "leaf length", "height","e/p"])
print("count:")
print(f.count())
print("mean:")
print(f.mean())
print("std:")
print(f.std())
print("quantile 0.25:")
print(f.quantile(0.25))
print("quantile 0.5:")
print(f.quantile(0.5))
print("quantile 0.75:")
print(f.quantile(0.75))
print("mode:")
print(f.mode('index'))
```
