Задание №3

```python
import pandas
import numpy

f = pandas.read_csv("travel-times.csv", engine = "python")
hasnan = []
avg = {}
for i in range(0,f.shape[1]):
    l = []
    hasnan.append(False)
    for j in range(0,f.shape[0]):
        if type(f.values[j][i]) == str:
            if f.values[j][i] == "-":
                hasnan[i] = True
            else:
                l.append(f.values[j][i])
        elif pandas.isnull(f.values[j][i]) == True:
            hasnan[i] = True
        else:
            l.append(f.values[j][i])
    if hasnan[i] == True:
        l = sorted(l)
        avg[f.columns[i]] = l[len(l)//2]
print(avg)
f = f.fillna(value=avg)
print(f)
```
